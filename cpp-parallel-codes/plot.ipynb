```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
m_data = np.loadtxt('m.txt')
m_data.shape
```

```python
x = np.arange(0, m_data.shape[0])
y = m_data
plt.plot(x,y,'b')
plt.xlabel('Number of Iterations')
plt.ylabel('Mean magnetisation <m>')
plt.text(10,0.1,r'T = 10')
plt.show()
```

```python
E_data = np.loadtxt('E.txt')
E_data.shape
```

```python
x = np.arange(0, E_data.shape[0])
y = E_data
plt.plot(x,y,'r')
plt.xlabel('Number of Iterations')
plt.ylabel('Energy E')
plt.show()
```

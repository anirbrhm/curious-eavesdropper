```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
m_data = np.loadtxt('m.txt')
m_data.shape
```

```python
x = np.arange(0, m_data.shape[0])
y = m_data
plt.plot(x,y,'b')
plt.xlabel('Number of Iterations')
plt.ylabel('Mean magnetisation <m>')
plt.show()
```

```python
E_data = np.loadtxt('E.txt')
E_data.shape
```

```python
x = np.arange(0, E_data.shape[0])
y = E_data
plt.plot(x,y,'r')
plt.xlabel('Number of Iterations')
plt.ylabel('Energy E')
plt.show()
```

```python
A_data = np.loadtxt('A.txt')
A_data.shape
```

```python
M = 100 
N = 100 
A_data = A_data.reshape(N+2,M+2)
A_data.shape
```

```python
COS = np.cos(A_data)
SIN = np.sin(A_data) 

X, Y = np.meshgrid(np.arange(0,A_data.shape[0]), np.arange(0,A_data.shape[1]))
plt.figure(figsize=(10,10), dpi=100)
Q = plt.quiver(X, Y, COS, SIN, units='width')
plt.axis('off')
plt.show()
```
